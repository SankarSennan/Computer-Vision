```python
from donut import DonutModel
from PIL import Image
import torch
model = DonutModel.from_pretrained("/content/donut1/result/train_sroie/train_sroie/20220830_042244")
if torch.cuda.is_available():
    model.half()
    device = torch.device("cuda")
    model.to(device)
else:
    model.encoder.to(torch.bfloat16)
model.eval()
image = Image.open("/content/donut1/sroie-donut/test/008.jpg").convert("RGB")
output = model.inference(image=image, prompt="<s_sroie-donut>")
output
```
